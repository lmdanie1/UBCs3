## Simplistic View of SDEs

Let $x^{\prime} = a x $ but then add some noise to the growth parameter $a$ to get a first order SDE.

Example 1: Add noise to the growth parameter $a$.
$$
    x^{\prime} = (a + \xi(t)) x \\
    dX = aXdt + X \xi(t)dt\\
    dX = aXdt +XdW(t)
$$

Example 2: Add noise to spatial coordinate. 
$$
    x^{\prime} = a(x+ \xi(t))\\
    dX = aXdt + a \xi(t)dt\\
    dX = aXdt +adW(t)
$$



## Visualize Noise

The term $x(t)$ is a continuously varying, normally distributed random variable. We can think of it discretely. Choose a time period $T>0$ and integer $N$.  

Let $\delta t = \frac{T}{N}$.  Then $xi(t_j)\in \mathcal{N}(\mu,\sigma^2)$ for each $t_j = j \delta t$. By "connecting the dots", we form a continuous function $\xi(t)$.

```python
##Visualize noise

N = 100 
noise = np.random.randn(N)
plt.plot(noise) 
plt.grid(True)
plt.ylim([-5,5])
plt.show()
```

## Discretized Brownian Motion

Aka Wiener Process

Choose time period $T>0$ and integrer $N$ and let $\delta t = T/N$.  A discretized Brownian moation is 
$$
W_0 = 0, W_j = W_{j-1}+dW_j ~\textrm{for} ~j = 1, ..,N
$$

where $dW_j \in \sqrt{\delta t} \mathcal{N}(0,1)$.

```python
T=1;
N=100;
dt= T/N;
sigma = np.sqrt(dt);
mu=0;
npaths = 5;
dW = sigma*np.random.randn(N,npaths) + mu
W = np.vstack([np.zeros(npaths),dW.cumsum(axis=0)])
t=np.linspace(0,T,N+1)
plt.plot(t,W,'b-', alpha=0.5)
plt.grid(True)
```

```python
## make it look more continuous by increasing N
T=1;
N=1000;
dt= T/N;
sigma = np.sqrt(dt);
mu=0;
npaths = 5;
dW = sigma*np.random.randn(N,npaths) + mu
W = np.vstack([np.zeros(npaths),dW.cumsum(axis=0)]) ##makes it start at zero instead of 1
t=np.linspace(0,T,N+1)
plt.plot(t,W, 'b-', alpha=0.5)
plt.grid(True)
```

## Stochastic integrals

$$

$$

## Numerical Methods SDEs

$$
X(t) = X_0 + \int_0^t f(X(s)) ds + \int_0^t g(X(s)) dW(s), 0 \leq t \leq T\\
dX(t) = f(X(t))dt + g(X(t)) dW(t),X(0) = X_0, 0 \leq t \leq T
$$
Let $\Delta t = \frac{T}{L} = R \delta t = R \frac{T}{N}$ so $L = N/R \in \mathbb{Z}$. Let $\tau_j = j \Delta t$ and $X_j = X(\tau_j)$.

$$
X_j = X_{j-1} + f(X_{j-1})\Delta \tau + g(X_{j-1})(W(\tau_j) - W(\tau_{j-1}))
$$

When $f(X) = aX$ and $g(X) = bX$, the exact solution is
$$
X(t) = X(0) \exp((a - b^2/2)t + bW(t)).
$$

```python
T=1;
N=1000;
dt= T/N;
R = 1;
dtau = R*dt;
sigma = np.sqrt(dt);
mu=0;
npaths = 1;
dW = sigma*np.random.randn(N,npaths) + mu
W = np.vstack([np.zeros(npaths),dW.cumsum(axis=0)])
t = np.linspace(0,T,N+1)
X = np.zeros(N+1)
a = 1;
b = 2;
X[0] = 1;
for j in range(1,N+1):
    X[j] = X[j-1] + a*X[j-1]*dtau + b*X[j-1]*dW[j-1]
    
plt.plot(X)

plt.show()
```

```python
T = 1; 
N = 1000; 
dt = T/N; 
R = 5; 
dtau = R*dt; 
L = N//R;
dW = np.sqrt(dt)*np.random.randn(N,1)
W = np.vstack([np.zeros(1),dW.cumsum(axis=0)])
t = np.linspace(0,T,L+1)
X = np.zeros(L+1)
a = 1; 
b = 2;
X[0] = 1
for j in range(1,L+1):
    X[j] = X[j-1] + a*X[j-1]*dtau + b*X[j-1]*(W[j*R] - W[(j-1)*R])
Xtrue = X[0]*np.exp((a - b**2/2)*t + b*W[range(0,N+1,R),:].flatten())
plt.plot(t,X,'b--',t,Xtrue,'r',alpha=0.5)

plt.show()
```
